```python
import matplotlib.pyplot as plt
import json
import numpy as np

with open("stats.txt", "r") as f:
    n_lines = sum(1 for _ in f)
    f.seek(0)
    for line in range(n_lines):
        line_data = json.loads(f.readline())
        data = np.array(line_data["eval_loss_stats"])
        plt.plot(data, label=f"{line_data["name"]} : {min(data)}")
    plt.yticks(np.arange(2, 13, 1))
    plt.grid()
    plt.title("eval_loss compare")
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import json
import numpy as np
import math

with open("stats.txt", "r") as f:
    lines = [json.loads(line) for line in f]

n_lines = len(lines)
n_cols = 5
n_rows = math.ceil(n_lines / n_cols)

plt.figure(figsize=(4 * n_cols, 3 * n_rows))

for i, line_data in enumerate(lines):
    plt.subplot(n_rows, n_cols, i + 1)
    eval_loss = line_data["eval_loss_stats"]
    train_loss = np.array(line_data["train_loss_stats"])
    plt.plot(eval_loss, label=f"eval (min={min(eval_loss):.3f})")
    plt.plot(train_loss, label=f"train (min={min(train_loss):.3f})")
    plt.legend()
    plt.title(line_data["name"])
    plt.yticks(np.arange(2, 13, 1))
    plt.grid()

plt.show()
```
