```python
import matplotlib.pyplot as plt
import json
import numpy as np

with open("stats.txt", "r") as f:
    n_lines = sum(1 for _ in f)
    f.seek(0)
    for line in range(n_lines):
        line_data = json.loads(f.readline())
        data = np.array(line_data["eval_loss_stats"])
        plt.plot(data, label=f"{line_data["name"]} : {min(data)}")
    plt.yticks(np.arange(2, 13, 1))
    plt.grid()
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import json
import numpy as np

with open("stats.txt", "r") as f:
    n_lines = sum(1 for _ in f)
    f.seek(0)
    for line in range(n_lines):
        line_data = json.loads(f.readline())
        data = np.array(line_data["eval_loss_stats"])
        plt.subplot(1, n_lines, line+1)
        plt.plot(data)
        plt.plot(np.array(line_data["train_loss_stats"]))
        plt.title(line_data["name"])
        plt.yticks(np.arange(2, 13, 1))
        plt.grid()
    plt.show()
```
